### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result
print(f'{establishments.count_documents(hygiene_query)} establishments rated Hygiene of 20')

# Display the first document in the results using pprint
pprint(establishments.find_one(hygiene_query))
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd
result = establishments.find(hygiene_query)
results = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
rows = len(results)
print(f'This data frame contains : {rows} rows. ')
# Display the first 10 rows of the DataFrame
results.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
rv_query = {'LocalAuthorityName': {'$regex':'London'},'RatingValue': {'$gte' : 4}}
# rv_query = {'RatingValue': {'$gte' : '4'}}
result = establishments.find(rv_query)
# # Use count_documents to display the number of documents in the result
print(f'{establishments.count_documents(rv_query)} documents found')

# Display the first document in the results using pprint
print(rv_query, result)
pprint(result[0])

```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.json_normalize(result)
# Display the number of rows in the DataFrame
print(f'{len(london_df)}')
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Latitude and longitude of Penang Flavours
pprint(establishments.find_one({'BusinessName': 'Penang Flavours'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {'RatingValue':5,
        '$and': [{'geocode.latitude': {'$gte': latitude - degree_search},
                  'geocode.latitude': {'$lt': latitude + degree_search},
                  'geocode.longitude': {'$gte': longitude - degree_search},
                  'geocode.longitude': {'$lte': longitude + degree_search}
            
       }]}
# Define the criteria
sort = [('scores.Hygiene', 1)]
# limit = 5
# pprint(list(establishments.find(query).sort(sort).limit(limit)))
# Print the results
query_results = list(establishments.find(query).sort(sort))
pprint(query_results[0:6])
```

```python
# Convert result to Pandas DataFrame
results_df = pd.json_normalize(query_results)
results_df 
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
#1. Matches establishments with a hygiene score of 0
hygiene_match = {'$match':{'scores.Hygiene': 0}}
# 2. Groups the matches by Local Authority
group_match = {'$group':{'_id': '$LocalAuthorityName', 'count':{'$sum': 1}}}
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort':{'count': -1}}

# Print the number of documents in the result
pipeline = [hygiene_match, group_match, sort_values]
results = list(establishments.aggregate(pipeline))
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
pipeline_db = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print(f'Number of rows" {len(pipeline_db)}')

# Display the first 10 rows of the DataFrame
pipeline_db.head(10)
```
